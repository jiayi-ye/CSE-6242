```python
def string_convert(input_string):
    if input_string == 'United Kingdom of Great Britain and Northern Ireland':
        return 'United Kingdom'
    elif input_string == 'Republic of Moldova':
        return 'Moldova, Republic of'
    elif input_string == 'Democratic Republic of the Congo':
        return 'Congo, the Democratic Republic of the'
    elif input_string == 'Bolivia (Plurinational State of)':
        return 'Bolivia, Plurinational State of'
    elif input_string == 'Venezuela (Bolivarian Republic of)':
        return 'Venezuela, Bolivarian Republic of'
    elif input_string == 'United Republic of Tanzania':
        return 'Tanzania, United Republic of'
    elif input_string == 'Democratic People\'s Republic of Korea':
        return 'Korea, Democratic People\'s Republic of'
    elif input_string == 'Republic of Korea':
        return 'Korea, Republic of'
    elif input_string == 'The former Yugoslav republic of Macedonia':
        return 'Macedonia, the former Yugoslav Republic of'
    elif input_string == 'United States of America':
        return 'United States'
    else:
        return input_string
```

```python
country_ids = {}
with open("data/world-110m-country-names.tsv") as fd:
    rd = csv.reader(fd, delimiter="\t")
    for row in rd:
        if row[1] == 'name':
            continue
        country_id = int(row[0])
        country_name = row[1]
        country_ids[country_name] = country_id
```

```python
dictionary = {}
```

```python
with open("data/tubercolusis_from 2007_WHO.csv") as fd:
    rd = csv.reader(fd, delimiter=",")
    first_row = None
    for row in rd:
        
        if row[0] == 'Country':
            first_row = row
            dictionary['countries'] = {}
            continue
            
        if string_convert(row[0]) not in dictionary['countries']:
            dictionary['countries'][string_convert(row[0])] = {}
            dictionary['countries'][string_convert(row[0])]['name'] = string_convert(row[0])
            if string_convert(row[0]) in country_ids:
                dictionary['countries'][string_convert(row[0])]['id'] = country_ids[string_convert(row[0])]
            else:
                dictionary['countries'][string_convert(row[0])]['id'] = -9999
                
        if 'time' not in dictionary['countries'][string_convert(row[0])]:
            dictionary['countries'][string_convert(row[0])]['time'] = {}

        dictionary['countries'][string_convert(row[0])]['time'][row[1]] = {}
        for i in range(2,len(first_row)):
            try:
                dictionary['countries'][string_convert(row[0])]['time'][row[1]][first_row[i]] = float(row[i].replace(' ', ''))
            except:
                if row[i] == '':
                    dictionary['countries'][string_convert(row[0])]['time'][row[1]][first_row[i]] = 0.0
                else:
                    dictionary['countries'][string_convert(row[0])]['time'][row[1]][first_row[i]] = float(row[i].split('-')[1])
```
